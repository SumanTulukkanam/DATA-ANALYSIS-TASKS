```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python

data=pd.read_csv("C:\\Users\\nseth\\Downloads\\householdtask3.csv")
```

```python
display(data.head(10))
```

```python
#Creating a scatter plot of year against own
plt.scatter(data['year'],data['own'])

#Adding title
plt.title('Scatter plot')
#Naming the axis
plt.xlabel('Year')
plt.ylabel('Age')

#Display graph
plt.show()
```

```python
#Creating a graph with year against own
plt.plot(data['year'])
plt.plot(data['own'])

#naming the plot
plt.title('Line Plot')

#Naming the axis
plt.xlabel('Year')
plt.ylabel('Own')


plt.show()
```

```python
#Ploting a bar chart
plt.bar(data['year'],data['own'])

#Adding Title
plt.title('Bar Plot')

#Naming axis
plt.xlabel('Year')
plt.ylabel('Own')

#display the plot
plt.show()

```

```python
# Creating a Histogram graph
plt.hist(data['income'])

#Naming the graph
plt.title("Histogram")

#display the graph
plt.show()
```
