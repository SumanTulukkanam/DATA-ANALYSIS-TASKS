```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv("C:\\Users\\suman\\Downloads\\householdtask3.csv")
```

```python
display(data.head(10))
```

```python
#Ploting a bar chart
plt.bar(data['year'],data['own'],color='blue')

#Adding Title
plt.title('Bar chart')

#labels
plt.xlabel('Year')
plt.ylabel('Own')

#Adding legends
plt.legend(['Own'])

#display the plot
plt.show()
```

```python
#Ploting a line chart
plt.plot(data['year'],  color='green')
plt.plot(data['own'], color='orange')

#Adding Title
plt.title('Line chart')

#Labels
plt.xlabel('Year')
plt.ylabel('Own')

#Adding legends
plt.legend(['Own'])
         
#display the plot
plt.show()
```
