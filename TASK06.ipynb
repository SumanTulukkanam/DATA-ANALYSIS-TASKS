```python
#Questions

#1. How many movies were released each year in netflix

#2. What is the range of IMDB score of movies in netflix

#3.How many genres is present in the dataset? How many movies each genre has?

#4. How does runtime of movies changed over the years?
```

```python
#Solution for Question 1

#1. How many movies were released each year in netflix


df["year"].value_counts()
```

```python
df.year.value_counts().plot(kind='bar', figsize=(6,5), color=('red','green','blue','yellow','violet','seagreen','orange','salmon'))
plt.xlabel('year')
plt.ylabel("Number of movies")
plt.title("Number of movies realsed in netfix, yearwise")
plt.show()
```

```python
#Solution for Question 2

#2. What is the range of IMDB score of movies in netflix



df.imdb_score.value_counts().plot(kind='bar',figsize=(16,8))

plt.xlabel('imdb_score')

plt.ylabel("Number of movies")

plt.title("IMDB score of movies")

plt.xticks(rotation=45)

plt.show()
```

```python
#Solution for Question 3

#3.How many genres is present in the dataset? How many movies each genre has?


num_genres = df['genre'].nunique()

print("Number of unique genres:", num_genres)
```

```python
df.genre.value_counts()
```

```python
df.genre.value_counts().plot(kind='line',figsize=(12,5))

plt.show()
```

```python
#Solution for Question 4

#4. How has runtime of movies changed over the years?

df.groupby('year')['runtime'].value_counts()
```

```python
df.groupby('year')['runtime'].nunique().plot(kind='bar',figsize=(12,5))

plt.title(" Movies of Varying runtime VS Year")

plt.xlabel("Year")

plt.ylabel("Count of Movies with different runtime")

plt.show()


```
